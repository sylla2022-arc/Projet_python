# Commentaire :
Les données générées par Faker ne permettent pas une dépendance mutuelle entre les colonnes. Pour rémédier à cela, il faudrait ajouter d'autre scripts.

Par exemple, ici le **salaire** n'est pas conforme  à la **fonction** correspondante. Il faudrait alors des scripts supplémenntaire pour les relier.

Ainsi, dans le seul but de fournir un tel traitement, je me restreins à un petit ensemble de données.


```python
df_copy  = df.sample(n= 10, random_state = 5, ignore_index = True)
```

```python
df_copy[['Job', 'Salary']].head(10)
```

```python
Fonction = {"spécialiste de l'accessibilité numérique":[20,30],'officier de police':[20,40],
                             'responsable des ressources humaines':[40,70],
                             'hydrogéologue':[25,30],'ingénieur logiciel':[30,35],
                             "juriste droit de l'environnement":[30,40],'assistant en architecture':[20,30],
                             "conseiller en assurances":[20,30],"ingénieur d'études en sûreté nucléaire":[70 ,100],
                             'développeur informatique':[45,65]}
salary = []                        
for val in Fonction.values():
    salary.append(round(random.randint(100*val[0],100*val[1])))
    
sal_zip = zip(df_copy['Job'].values.tolist(),salary )
```

```python
df_trans = pd.DataFrame(sal_zip , columns = ["Job", 'Salary'])
df_trans
```

## Actualisation

```python
del  df_copy['Salary']
```

```python
df_final = pd.merge(df_copy, df_trans , how = 'inner', on = 'Job')
sal = df_final.pop('Salary')
df_final.insert(9, 'Salary', sal)
```

```python
df_final.head(4)
```

## Commentaire :

Nous venons de voir une façon de lier les colonnes entre elles, on voit bien à quel point cela peut être fastidieux sur certain type de données.
